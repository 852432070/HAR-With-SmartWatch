```python
import re
import matplotlib.pyplot as plt

def extract_val_accuracies(log_file_path):
    with open(log_file_path, 'r') as file:
        log_content = file.read()
    pattern = re.compile(r'val_accuracy: (\d+\.\d+)')
    matches = re.findall(pattern, log_content)
    return [float(match) for match in matches]

# 定义要解析的文件列表
log_files = [
    "/data/wang_sc/HAR-With-SmartWatch/attention_based/log_LIMU_self_made.txt",
    "/data/wang_sc/HAR-With-SmartWatch/attention_based/log_sensor_self_made.txt",
    "/data/wang_sc/HAR-With-SmartWatch/attention_based/log_SE_self_made.txt"
    # 更多文件路径...
]

log_labels = [
    "LIMUNet", "Sensor Attention", "Channel Attention"
]

# 为每个文件绘制一条折线
for i in range(len(log_files)):
    val_accuracies = extract_val_accuracies(log_files[i])
    plt.plot(val_accuracies, linestyle='-', label=log_labels[i])

# 设置标题和轴标签
plt.title('Train Process')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')

# 开启网格和图例
plt.grid(True)
plt.legend()

# 显示图表
plt.show()
```
